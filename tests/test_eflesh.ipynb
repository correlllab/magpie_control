```python
from magpie_control.gripper import Gripper

# This will print port detection results
gripper = Gripper(debug=False)
print(f"Eflesh available: {gripper.eflesh_available}")
print(f"Number of sensors: {len(gripper.eflesh_sensors)}")
```

```python
gripper.reset_parameters()
```

```python
# Test with dynamixel load sensing
result1 = gripper.deligrasp(x=50, fc=5.0, dx=2, df=0.5)
```

```python
# Test with eflesh sensing
result2 = gripper.deligrasp_eflesh(x=50, fc=5.0, dx=2, df=0.5, num_samples=1)
```

```python
# Compare logs
print("Dynamixel log:", len(result1[3]), "entries")
print("Eflesh log:", len(result2[3]), "entries")
```
